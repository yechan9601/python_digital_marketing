reset_index메서드는 설정 인덱스를 제거하고 기본 인덱스(0,1,2, ... , n)으로 변경하는 메서드 입니다.

```python
date_month_map = {
    str(x)[:10]: 'M_%s' % (i+1) for i, x in enumerate(
        sorted(data_df.reset_index()['InvoiceDate'].unique(), reverse=True)
    )
}
```

- sorted(data_df.reset_index()['InvoiceDate'].unique() => 고유한 InvoiceDate를 내림차순으로 정렬

- for i, x in enumerate() => i는 순번(0, 1, 2...), x에 InvoiceDate

- str(x)[:10]: 'M_%s' % (i+1) => 날짜타입인 x를 String으로 변환, 이를 'M_순번' 형태로 딕셔너리 저장

# Recency, frequency, monetary

[RFM분석이란 무엇일까요?](https://www.datarian.io/blog/what-is-rfm)

```python
today_date = dt.datetime(2011, 12, 11)
```

```python
rfm = df.groupby('CustomerID').agg(
    {
      'InvoiceDate': [
        lambda date: (date.max() - date.min()).days,
        lambda date: (today_date - date.min()).days
      ], 
     'InvoiceNo': lambda num: num.nunique(),
     'Sales': lambda TotalPrice: TotalPrice.sum()
    }
)

rfm.columns = rfm.columns.droplevel(0)
rfm.columns = ['recency_cltv_p', 'T', 'frequency', 'monetary']
rfm.head()
```

```python
rfm['monetary'] = rfm['monetary'] / rfm["frequency"]

# monetary를 frequency로 나누어주었으니, 'monetary 평균'으로 컬럼명 바꾸기
rfm.rename(columns={"monetary": "monetary_avg"}, inplace=True) 

rfm["recency_weekly_p"] = rfm["recency_cltv_p"] / 7
rfm["T_weekly"] = rfm["T"] / 7

rfm = rfm[rfm["monetary_avg"] > 0]
rfm = rfm[(rfm['frequency'] > 1)] # at least one purchase should be made
rfm["frequency"] = rfm["frequency"].astype(int)

rfm.head()
```

# BG/NBD Model

고객의 예상 구매 횟수를 계산하는 BG/NBD 모형

1. 고객이 남아있는 동안, 일정한 기간 T 동안의 구매 횟수는 Pois(λT)를 따릅니다. 1일 간 Pois (1/12)를 따른다면 T= 1년일 경우 Pois (30)을 따르게 되겠죠! (포아송의 모수인 λT는 예상 구매 횟수 를 의미합니다)

2. 고객마다 일정한 기간 동안 구매하는 횟수는 다릅니다. 이는 λ ~ Gamma (r, α)을 따릅니다.

3. j번째 구매가 마지막이고 더 이상 구매를 하지 않을 확률 (이탈률)은 p입니다. 이탈할 때까지의 구매 횟수는 Geo (p)를 따릅니다.

4. 고객마다 더 이상 구매를 하지 않을 확률 (이탈률)은 다릅니다. 이탈률 p는 p ~ Beta (a,b)를 따릅니다.

5. 고객별 일정 기간 동안의 구매 횟수와 구매를 하지 않을 확률은 서로 영향을 주지 않습니다.

```python
bgf = BetaGeoFitter(penalizer_coef=0.001)

bgf.fit(rfm['frequency'],
        rfm['recency_weekly_p'],
        rfm['T_weekly'])
```

# Gamma-Gamma Model

1.고객별 구매 금액은 평균 구매 금액을 중심으로 랜덤하게 분포합니다.

2.고객들의 평균 구매 금액은 고객마다 다르지만, 한 고객의 평균 구매 금액은 시간에 따라 불변합니다.

3.이 평균 구매 금액은 구매 과정과 독립이여야 합니다.

```python
ggf = GammaGammaFitter(penalizer_coef=0.01)
ggf.fit(rfm['frequency'], rfm['monetary_avg'])

ggf.conditional_expected_average_profit(rfm['frequency'],
                                        rfm['monetary_avg']).head(10)

rfm["expected_average_profit"] = ggf.conditional_expected_average_profit(rfm['frequency'], rfm['monetary_avg'])

rfm.sort_values("expected_average_profit", ascending=False).head(10)
```

1 Month CLV Prediction with BG-NBD && GG Model

```python
cltv_1 = ggf.customer_lifetime_value(bgf,
                                     rfm['frequency'],
                                     rfm['recency_weekly_p'],
                                     rfm['T_weekly'],
                                     rfm['monetary_avg'],
                                     time=1, # 1 month
                                     freq='W', # Tenure frequency information-weekly.
                                     discount_rate=0.01)
                                     
cltv_1.head()
```

```python
cltv_1 = cltv_1.reset_index()
cltv_1.rename(columns={"clv":"clv_1"}, inplace=True)
cltv_1.head()
```

3 Months CLV Prediction with BG-NBD & GG Model

```python
cltv_3 = ggf.customer_lifetime_value(bgf,
                                     rfm['frequency'],
                                     rfm['recency_weekly_p'],
                                     rfm['T_weekly'],
                                     rfm['monetary_avg'],
                                     time=3, # 1 month
                                     freq='W', # Tenure frequency information-weekly.
                                     discount_rate=0.01)

cltv_3 = cltv_3.reset_index()
cltv_3.rename(columns={"clv":"clv_3"}, inplace=True)
cltv_3.head()
```
